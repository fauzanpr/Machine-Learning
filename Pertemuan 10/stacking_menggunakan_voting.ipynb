# Stacking menggunakan Metode Voting

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import VotingClassifier # import model Voting
from sklearn.ensemble import RandomForestClassifier # import RF
from sklearn.ensemble import AdaBoostClassifier # import AdaBoost
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python
df_rf = pd.read_csv('data/mushrooms.csv')
df_ad = pd.read_csv('data/mushrooms.csv')
```

# Random Forest

## Persiapan Data

### Membuat X dan y

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Inisiasi label encoder
encode = LabelEncoder()

# Terpakan label encoder
df_rf['cap-shape'] = encode.fit_transform(df_rf['cap-shape'])
df_rf['cap-surface'] = encode.fit_transform(df_rf['cap-surface'])
df_rf['cap-color'] = encode.fit_transform(df_rf['cap-color'])
df_rf['bruises'] = encode.fit_transform(df_rf['bruises'])
df_rf['odor'] = encode.fit_transform(df_rf['odor'])
df_rf['gill-attachment'] = encode.fit_transform(df_rf['gill-attachment'])
df_rf['gill-spacing'] = encode.fit_transform(df_rf['gill-spacing'])
df_rf['gill-size'] = encode.fit_transform(df_rf['gill-size'])
df_rf['gill-color'] = encode.fit_transform(df_rf['gill-color'])
df_rf['stalk-shape'] = encode.fit_transform(df_rf['stalk-shape'])
df_rf['stalk-root'] = encode.fit_transform(df_rf['stalk-root'])
df_rf['stalk-surface-above-ring'] = encode.fit_transform(df_rf['stalk-surface-above-ring'])
df_rf['stalk-surface-below-ring'] = encode.fit_transform(df_rf['stalk-surface-below-ring'])
df_rf['stalk-color-above-ring'] = encode.fit_transform(df_rf['stalk-color-above-ring'])
df_rf['stalk-color-below-ring'] = encode.fit_transform(df_rf['stalk-color-below-ring'])
df_rf['veil-type'] = encode.fit_transform(df_rf['veil-type'])
df_rf['veil-color'] = encode.fit_transform(df_rf['veil-color'])
df_rf['ring-number'] = encode.fit_transform(df_rf['ring-number'])
df_rf['ring-type'] = encode.fit_transform(df_rf['ring-type'])
df_rf['spore-print-color'] = encode.fit_transform(df_rf['spore-print-color'])
df_rf['population'] = encode.fit_transform(df_rf['population'])
df_rf['habitat'] = encode.fit_transform(df_rf['habitat'])
```

```python
df_rf
```

```python
# Seleksi fitur

# Slice dataframe
X_rf = df_rf.iloc[:, 1:-1]
y_rf = df_rf['class']
y_rf = y_rf.map({'e':1, 'p':0}) # Encode label

# Cek jumlah fitur dan instance
X_rf.shape
```

```python
y_rf
```

```python
from sklearn.model_selection import train_test_split

X_rf_train, X_rf_test, y_rf_train, y_rf_test = train_test_split(X_rf, y_rf, test_size=0.2, random_state=1)
```

### Proses Training

```python
dt_rf = RandomForestClassifier(n_estimators=100, random_state=1)

# Sesuaikan dt ke set training
dt_rf.fit(X_rf_train, y_rf_train)
```

```python
# Memprediksi label set test
y_pred_rf = dt_rf.predict(X_rf_test)

#  menghitung set accuracy
acc_rf = accuracy_score(y_rf_test, y_pred_rf)
print("Test set accuracy: {:.2f}".format(acc_rf))
print(f"Test set accuracy: {acc_rf}")
```

# Stacking menggunakan Voting

```python
# Definisikan algoritma yang akan digunakan untuk voting

clf1 = RandomForestClassifier()
clf2 = AdaBoostClassifier()

# model hard voting
voting = VotingClassifier(estimators=[('RandomForestClassifier', clf1), ('AdaBoostClassifier', clf2)], voting='hard')
voting2 = VotingClassifier(estimators=[('RandomForestClassifier', clf1), ('AdaBoostClassifier', clf2)], voting='hard')
```

```python
# Fit model
voting.fit(X_rf_train, y_rf_train)

# Prediksi
y_pred_vt1 = voting.predict(X_rf_test)

# Evaluasi akurasi testing data
acc_vt1 = accuracy_score(y_rf_test, y_pred_vt1)

# Print hasil evaluasi
print('Voting Hard')
print("Test set accuracy: {:.2f}".format(acc_vt1))
print(f"Test set accuracy: {acc_vt1}")
```

```python
# Fit model
voting2.fit(X_ad_train, y_ad_train)

# Prediksi
y_pred_vt2 = voting2.predict(X_ad_test)

# Evaluasi akurasi testing data
acc_vt2 = accuracy_score(y_ad_test, y_pred_vt2)

# Print hasil evaluasi
print('Voting Hard')
print("Test set accuracy: {:.2f}".format(acc_vt2))
print(f"Test set accuracy: {acc_vt2}")
```
